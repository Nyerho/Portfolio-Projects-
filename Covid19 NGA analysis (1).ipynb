## What was the mortality rate in NGA from covid19

```python
# mortality rate is number of deaths/ number of confirmed 
mortality = df.groupby('date')[['confirmed', 'dead']].sum()
mortality = pd.DataFrame(mortality)

#create mortality rate column
mortality['mortality_rate%'] = round(mortality['dead']/mortality['confirmed']*100,2)

# For columns where both values are 0, the result will be a NaN value, so I will replace NaN with 0
mortality['mortality_rate%'] = mortality['mortality_rate%'].fillna(0)

# 0 divided by a whole number eg (0/1) will result in inf. So we I replace inf value with 0
mortality['mortality_rate%'] = mortality['mortality_rate%'].replace(np.inf, 0)

# Show the result
mortality.sample(10)
```

```python
## create fig and axis object 
fig,ax = plt.subplots(figsize = (12,6))

# create plot to show time series for all cases 
plt.plot(mortality.index,mortality['mortality_rate%'], color = 'black')

# set the other parameters
plt.title('Line plot showing trend of mortality rate')
plt.xlabel('date')
plt.xticks(rotation = 90)
plt.ylabel('Mortality rate %')

plt.show()
```

```python
answer4 = round((answer2/answer1)*100,2)
print('Total number of mortality rate in Nigeria as at the time of this analysis is {}%' .format(answer4))
```

## Q5. what was the recovery rate in nigeria

```python
## create fig and axis object 
fig,ax = plt.subplots(figsize = (12,6))

# create plot to show time series for all cases 
plt.plot(recovery.index,recovery['recovery%'], color = 'brown')

# set the other parameters
plt.title('Line plot showing trend of Recovery rate')
plt.xlabel('date')
plt.xticks(rotation = 90)
plt.ylabel('Recovery rate %')

plt.show()
```

## Q6. What was the daily average number of confirmed, death and recovered cases in Nigeria?

```python
## find daily avg of all cases and group by date 
avg_cases = df.groupby('date')[['confirmed', 'dead', 'recovered']].mean()
avg_cases = round(avg_cases)

#convert result to dataframe and create column for avg_cases 

avg_cases = pd.DataFrame(avg_cases)
avg_cases.sample(10)
```

```python
# create fig and axis label 
fig, ax = plt.subplots(figsize = (10,6))

# create plot to show trends for all cases 
plt.plot(avg_cases.index, avg_cases['confirmed'], color = 'blue', label = 'Daily average of confirmed cases')
plt.plot(avg_cases.index, avg_cases['dead'], color = 'red', label = 'Daily average of dead cases')
plt.plot(avg_cases.index, avg_cases['recovered'], color = 'green', label = 'Daily average of recovered cases')


# set parameters
plt.legend()
plt.xlabel('Date')
plt.xticks(rotation = 90)
plt.ylabel('Average number of cases')
plt.title('Line plot showing average confirmed dead and recoverd cases in Nigeria')

plt.show()
```

```python
answer6 =  round(avg_cases['confirmed'].mean())
answer7 =  round(avg_cases['dead'].mean(),2)
answer8 =  round(avg_cases['recovered'].mean())

print('The daily average number of confirmed cases in Nigeria was {}'.format(answer6))
print('The daily average number of death cases in Nigeria was {}'.format(answer7))
print('The daily average number of recovered cases in Nigeria was {}'.format(answer8))
```

## Q7. what was the monthly avg number confirmed, dead and recovered cases in Nigeria?

```python
# group all 3 case by same month and year, then visualize these cases using a bar plot
```

```python
#create a column that has only month and year values 
df['month_year'] = df['date'].dt.strftime('%m-%Y')

# the result of this has a dtype 'O' to convert to 'datetype' calendar format
df['month_year'] = pd.to_datetime(df['date']).dt.to_period('M')

df['month_year'].dtype
```

```python
# Calculate the average sales for each product for each month.
monthly_avg_cases = df.groupby('month_year')[['confirmed', 'dead', 'recovered']].mean()
# Combine the average sales for each product into a single table.
monthly_avg_cases = monthly_avg_cases.reset_index()

#convert result to dataframe and create month_year column
monthly_avg_cases = pd.DataFrame(monthly_avg_cases) 

#round up the values for confirmed and recovered but leave death cases because rounding it will give 0 for all values
monthly_avg_cases[['confirmed','recovered']] = round(monthly_avg_cases[['confirmed','recovered']])



# Print the monthly average sales for each product.
monthly_avg_cases
```

```python
## plot bar chart to visualize data result
for cases,title in zip(['confirmed','dead','recovered'], ['Bar Plot showing monthly average confirmed cases in Nigeria from 2020-2022', 'Bar Plot showing monthly average death cases in Nigeria from 2020-2022', 'Bar Plot showing monthly average recovered cases in Nigeria from 2020-2022']):
    
    # Create a figure and axis object 
    fig, ax = plt.subplots(figsize = (10,6))
    
    # Create a bar plot for all cases
    sns.barplot(x = monthly_avg_cases.index, y = monthly_avg_cases[cases])
    
    # Rotate the xlabel to easier readability
    plt.xticks(rotation = 90)
   
    # Create the title for all cases
    plt.title(title)
    
    plt.show()
    
```

```python
answer9 =   round(monthly_avg_cases['confirmed'].mean())
answer10 =  round(monthly_avg_cases['dead'].mean(),2)
answer11 =  round(monthly_avg_cases['recovered'].mean())

print('Total monthly average number of confirmed cases in Nigeria is {}'.format(answer9))
print('Total monthly average number of dead cases in Nigeria is {}'.format(answer10))
print('Total monthly average number of recovered cases in Nigeria {}'.format(answer11))
```

## what state was most affected most by the pandemic 

```python
## group cases by state and the visualise result with bar plot 
state = df.groupby('state')[['confirmed', 'dead', 'recovered']].sum()

# create df to hold this info
state = pd.DataFrame(state)

state.head(10)
```

```python
## visualize using a bar plot
#arrang chart in descending order
state = state.sort_values(by = 'confirmed', ascending = False)

#create fig and axis 
fig,axs = plt.subplots(1,3, figsize = (15,6))


# plot bar for confirmed cases
state1 = sns.barplot(data = state, x = state.index, y = state.confirmed, ax = axs[0])
state1.set(title = 'Confirmed cases per state')
state1.set_xticklabels(state1.get_xticklabels(), rotation = 90, size = 10)


# plot bar for dead cases
state1 = sns.barplot(data = state, x = state.index, y = state.dead, ax = axs[1])
state1.set(title = 'Dead cases per state')
state1.set_xticklabels(state1.get_xticklabels(), rotation = 90, size = 10)


# plot bar for recovered cases
state1 = sns.barplot(data = state, x = state.index, y = state.recovered, ax = axs[2])
state1.set(title = 'Recovered cases per state')
state1.set_xticklabels(state1.get_xticklabels(), rotation = 90, size = 10)

# Set spacing between suplots
fig.tight_layout()

# Display the bar chart
plt.show()
```

```python
##create a piechart to show top 10 affected states 
# Arrange in descending order
top10states = state.sort_values(by = 'confirmed', ascending = False)[:10]

# Create a figure with three subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 5))

# Create the pie charts for confirmed subplot
axs[0].pie(top10states['confirmed'], labels=top10states.index, autopct='%1.1f%%')
axs[0].set_title('Confirmed Cases')

# Create the pie charts for death subplot
axs[1].pie(top10states['dead'], labels=top10states.index, autopct='%1.1f%%')
axs[1].set_title('Dead Cases')

# Create the pie charts for recovered subplot
axs[2].pie(top10states['recovered'], labels=top10states.index, autopct='%1.1f%%')
axs[2].set_title('Recovered Cases')

# Display the pie charts
plt.show()
```

#### result from this analysis shows that Lagos and FCT were the most affected states in comparison to the other states, hence they are labeled High Risk Areas
